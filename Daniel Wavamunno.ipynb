```python
flood['target'] = flood.FloodProbability

x = flood[['MonsoonIntensity','TopographyDrainage','RiverManagement','Deforestation','Urbanization','ClimateChange','DamsQuality',
           'Siltation','AgriculturalPractices','Encroachments','IneffectiveDisasterPreparedness','DrainageSystems','CoastalVulnerability',
           'Landslides','Watersheds','DeterioratingInfrastructure','PopulationScore','WetlandLoss','InadequatePlanning','PoliticalFactors']]
y = flood['target']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 42)
```

```python
## Support Vector Regressor
from sklearn.svm import SVR

model=SVR()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error

correlation_score = r2_score(y_test,y_pred)
rms_error = root_mean_squared_error(y_test,y_pred)
```

```python
print(f'correlation score (R2_values) : {correlation_score}\n')
print(f"Error: {rms_error} \n")
```

```python
## Decison Tree Regressor
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error

correlation_score = r2_score(y_test,y_pred)
```

```python
print(f'correlation score (R2_values) : {correlation_score}\n')
```

```python
## Linear Regression
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = root_mean_squared_error(y_test,y_pred)
correlation_score = r2_score(y_test, y_pred)

print(f"RÂ² score : {correlation_score}\n Mean Squared Error  : {mse}")
```

## Hyper Parameter for 'Tuning Support Vector Regressor' and 'Decision Tree Regressor'

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'max_depth' : [3, 5, 7, 10],
    'min_samples_split' : [2, 5, 10],
    'min_samples_leaf' : [1, 2, 5],
    'criterion' : ['absolute_error','squared_error','poisson','friedman_mse']
}

grid_search = GridSearchCV(
    estimator = DecisionTreeRegressor(random_state=42),
    param_grid = param_grid,
    scoring = 'neg_mean_squared_error',
    cv = 5,
    n_jobs = -1
)
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Initialize the model
dtr = DecisionTreeRegressor(
    criterion='squared_error',
    max_depth=10,
    min_samples_leaf=5,
    min_samples_split=2
)

# Fit the model
dtr.fit(x_train, y_train)

# Make predictions
y_pred = dtr.predict(x_test)

# Calculate R^2 score
r2 = r2_score(y_test, y_pred)  # Renamed to 'r2' to avoid conflict

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))  # Use mean_squared_error with sqrt for RMSE

print("R^2 Score:", r2)
print("Root Mean Squared Error:", rmse)
```

## SVR hyper parameter tuning

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for SVR
param_grid = {
    'C': [0.1, 1, 10, 100],              # Regularization parameter
    'epsilon': [0.1, 0.2, 0.5, 1],       # Epsilon in the epsilon-SVR model
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],  # Kernel types
    'degree': [2, 3, 4]                  # Degree of the polynomial kernel (only applies if kernel='poly')
}

# Set up GridSearchCV with SVR
grid_search = GridSearchCV(
    estimator=SVR(),
    param_grid=param_grid,
    scoring='neg_mean_squared_error',
    cv=5,
    n_jobs=-1
)
```

```python
# Fit the grid search to the data
grid_search.fit(x_train, y_train)
```

```python
# best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Score (neg MSE):", best_score)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Get the best parameters from the grid search
best_params = grid_search.best_params_

# Initialize SVR with the best parameters
best_svr = SVR(**best_params)

# Fit the model to the training data
best_svr.fit(x_train, y_train)

# Make predictions on the test data
y_pred = best_svr.predict(x_test)

# Calculate RÂ² score
r2 = r2_score(y_test, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("RÂ² Score:", r2)
print("Root Mean Squared Error:", rmse)
```
